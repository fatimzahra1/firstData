## The best and the worst movies

```python
from IPython.display import HTML
pd.options.display.max_colwidth=200
```

```python
df_best=df[["poster_path","title","budget_musd","revenue_musd","vote_count","vote_average","popularity"]].copy()
df_best
```

```python
df_best["profit_musd"]=df.revenue_musd.sub(df.budget_musd)

df_best["return"]=df.revenue_musd.div(df.budget_musd)
```

```python
df_best
```

```python
df_best.columns = ["", "Title", "Budget", "Revenue", "Votes", 
                   "Average Rating", "Popularity", "Profit", "ROI"]
```

```python
df_best
```

```python
df_best.set_index("Title",  inplace = True)
```

```python
df_best
```

```python
df_best.iloc[0,0]
```

```python
subset=df_best.iloc[:5,:2]
subset
```

```python
HTML(subset.to_html(escape=False))
```

```python
df_best.sort_values(by="Average Rating", ascending=False)
```

```python
df_best.sort_values(by="ROI",ascending=False)
```

```python
df_best.loc[df_best.Budget>=5].sort_values(by="ROI",ascending=False)
```

```python
df_best.Budget.fillna(0,inplace=True)
df_best.Votes.fillna(0,inplace=True)
```

```python
df_best.info()
```

```python
def best_worst(n,by,min_budg=0,min_votes=0, ascending=False):
    df2=df_best.loc[(df_best.Budget>=min_budg)&(df_best.Votes>=min_votes)].sort_values(by=by, ascending=ascending).head(n).copy()
    return HTML(df2.to_html(escape=False))
```

## Top 5 movies -Highest Revenues

```python
df1=best_worst(n=5, by="Revenue")
```

```python
df1
```

```python
best_worst(n=5, by="Profit")
```

```python
best_worst(n=5, by="Profit", ascending=True)
```

```python
best_worst(n=5, by="ROI", min_budg=50)
```
